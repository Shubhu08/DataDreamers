```python
data_Geographic_region_specific2.hist(column="TPopulation1July")
```

```python
data_Geographic_region_specific2['TPopulation1July']=(data_Geographic_region_specific2['TPopulation1July']-data_Geographic_region_specific2['TPopulation1July'].mean())/data_Geographic_region_specific2['TPopulation1July'].std()
```

```python
data_Geographic_region_specific2.hist(column="TPopulation1July")
```

```python
data_Geographic_region_specific2.head()
```

```python
data_Geographic_region_specific3 = data_Geographic_region_specific2
```

```python
from numpy import percentile
#calculate interquartile range as the data distribution is skewed
Q1 = data_Geographic_region_specific3.quantile(0.25)
Q3 = data_Geographic_region_specific3.quantile(0.75)
IQR = Q3 - Q1
data_Geographic_region_specific4 = data_Geographic_region_specific3[~((data_Geographic_region_specific3 < (Q1 - 1.5 * IQR)) |(data_Geographic_region_specific3 > (Q3 + 1.5 * IQR))).any(axis=1)]
data_Geographic_region_specific4.shape
```

```python
data_Geographic_region_specific4.hist(column="TPopulation1July")
```

```python
data_Geographic_region_specific4.hist(bins=50, figsize=(12,9))
plt.show()
```

```python
#Checking for the p-value of the features
import statsmodels.api as sm
import pandas as pd

# defining the dependent and independent variables
Xtrain = data_Geographic_region_specific4.drop('NetMigrations', axis = 1)
Ytrain = data_Geographic_region_specific4['NetMigrations']

# building the model and fitting the data
lin_reg = sm.OLS(Ytrain, Xtrain).fit()
print(lin_reg.summary())
```
