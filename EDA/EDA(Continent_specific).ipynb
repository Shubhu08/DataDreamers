```python
y_continent = data_Geographic_region_specific2['Location']
x_male = data_Geographic_region_specific2['DeathsMale']
x_female = data_Geographic_region_specific2['DeathsFemale'] * -1
```

```python
# Creating instance of the figure
configure_plotly_browser_state()
fig = gp.Figure()
  
# Adding Migrants Enter data to the figure
fig.add_trace(gp.Bar(y= y_continent, x = x_male, 
                     name = 'Male Mortality', 
                     orientation = 'h'))
  
# Adding Migrants Leave data to the figure
fig.add_trace(gp.Bar(y = y_continent, x = x_female,
                     name = 'Female Mortality', orientation = 'h'))
  
# Updating the layout for our graph
fig.update_layout(title = 'Interactive Population Pyramid of Mortality (1960-2020)',
                 title_font_size = 22, barmode = 'relative',
                 bargap = 0.0, bargroupgap = 0,
                 xaxis = dict(tickvals = [-600000000, -400000000, -200000000,
                                          0, 200000000, 400000000, 600000000],
                                
                              ticktext = ['6M', '4M', '2M', '0', 
                                          '2M', '4M', '6M'],
                                
                              title = 'Population in Millions',
                              title_font_size = 14)
                 )
  
fig.show()
```
