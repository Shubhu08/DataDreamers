```python
df = df[['Location', 'Time', 'Births', 'TPopulation1July', 'PopDensity', 'PopSexRatio', 'LEx', 'Deaths', 'InfantDeaths', 'IMR', 'NetMigrations', 'Unemployment', 'GDP']]
df['GDP'] = df['GDP'].fillna(df['GDP'].median())
df['Unemployment'] = df['Unemployment'].fillna(df['Unemployment'].median())

# Missing Value Interpolation
def interpolate(d, on='date', vals=['value']):
    return (d.set_index(on).reindex(pd.date_range(d[on].min(), d[on].max()))
             [vals].interpolate()
             .rename_axis(on)
           ) 

# df.groupby('cat').apply(interpolate).reset_index()
# df = interpolate(df, on='Time', vals=['TPopulation1July', 'PopDensity', 'PopSexRatio', 'LEx', 'Deaths', 'InfantDeaths', 'IMR', 'NetMigrations', 'Unemployment', 'GDP'])
# df = df.replace(np.nan, 0)

df.head()
```

```python
import matplotlib.pyplot as plt

df.hist(bins=50, figsize=(20,10))
plt.show()
```

```python
# Fix skewed variables by logging
for var in ['TPopulation1July', 'Deaths', 'Births', 'InfantDeaths', 'PopDensity']:
  df[var] = np.log(df[var])
```

```python
df.hist(bins=50, figsize=(20,10))
plt.show()
```

```python
top10_countries = df.groupby('Location')['NetMigrations'].sum().nlargest(5)
# median10_countries = df.groupby('Location')['NetMigrations'].median().nlargest(3)
bottom10_countries = df.groupby('Location')['NetMigrations'].sum().nsmallest(5)
df_choosenCountry_10 = pd.concat([top10_countries, bottom10_countries])

df_choosenCountry_10 = df.loc[df['Location'].isin(list(df_choosenCountry_10.index))]
df_choosenCountry_10
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(11.7,8.27))

data = df_choosenCountry_10

sns.lineplot(x='Time', y='PopDensity', hue='Location',data=data, ci=None, ax=ax[0])
sns.move_legend(ax[0], "upper left", bbox_to_anchor=(1.25, 1)) 

sns.lineplot(x='Time', y='Births', hue='Location', data=data, ci=None, ax=ax[1]) 
ax[1].get_legend().remove()

sns.lineplot(x='Time', y='Deaths', hue='Location',data=data, ci=None, ax=ax[2]) 
ax[2].get_legend().remove() 

sns.lineplot(x='Time', y='GDP', hue='Location',data=data, ci=None, ax=ax[3]) 
ax[3].get_legend().remove()

sns.lineplot(x='Time', y='Unemployment', hue='Location',data=data, ci=None, ax=ax[4]) 
ax[4].get_legend().remove()
```
